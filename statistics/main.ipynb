```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Reading data from csv

```python
data = pd.read_csv("../technologies.csv")
data["english"].replace(np.nan, "No info", inplace=True)
data["datetime"] = pd.to_datetime(data["datetime"])
data.info()
```

# The most demanded technologies
This plot shows technologies that occured in the vacancy descriptions the most for any experience.

```python
technology_counts = data["technology"].value_counts()

plt.figure(figsize=(14, 6))
technology_counts.plot(kind='bar', color='skyblue')
plt.title('Technology Occurrence Frequency')
plt.xlabel('Technology')
plt.ylabel('Occurence count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# The most demanded technologies for 1 year or less experience
Here we can see the must-have skills for Junior Python devs.

```python
filtered_data = data[data["experience"] <= 1]

technology_counts = filtered_data["technology"].value_counts()

plt.figure(figsize=(14, 6))
technology_counts.plot(kind='bar', color='skyblue')
plt.title('1 year or less experience Technology Occurrence Frequency')
plt.xlabel('Technology')
plt.ylabel('Occurence count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# The most demanded technologies for 2-4 years experience
Now we will analyze the demanded skills for Middle Python devs.

```python
filtered_data = data[(data["experience"] > 1) & (data["experience"] < 5)]

technology_counts = filtered_data["technology"].value_counts()

plt.figure(figsize=(14, 6))
technology_counts.plot(kind='bar', color='skyblue')
plt.title('2-4 years experience Technology Occurrence Frequency')
plt.xlabel('Technology')
plt.ylabel('Occurence count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# The most demanded technologies for 5+ years experience
And finally we can see what skills are required for Seniors.

```python
filtered_data = data[data["experience"] > 4]

technology_counts = filtered_data["technology"].value_counts()

plt.figure(figsize=(14, 6))
technology_counts.plot(kind='bar', color='skyblue')
plt.title('2-4 years experience Technology Occurrence Frequency')
plt.xlabel('Technology')
plt.ylabel('Occurence count')
plt.xticks(rotation=45, ha='right')
plt.show()
```
